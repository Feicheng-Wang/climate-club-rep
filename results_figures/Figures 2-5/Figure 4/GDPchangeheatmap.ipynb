```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['Simsun']
plt.rcParams['axes.unicode_minus'] = False

# 文件路径
file_path = 'clubresults0322.xlsx'
sheet_name = 'C_GDP'

# 读取数据（B1:K8，A列是行标签）
df = pd.read_excel(file_path, sheet_name=sheet_name, index_col=0, usecols="A:K", nrows=7)

# 保留两位小数
df = df.round(2)

# 设定国家分类颜色（你可以根据政策覆盖逻辑自定义颜色）
column_colors = {
    'EU': 'goldenrod',
    'ANNEX I': 'deepskyblue',
    'CAN': 'deepskyblue',
    'USA': 'deepskyblue',
    'MEX': 'firebrick',
    'CHN': 'firebrick',
    'IND': 'firebrick',
    'ROW': 'gray',
    'RUS': 'gray',
    'OILEX': 'gray'
}

# 提取列标题颜色（按顺序）
col_colors = [column_colors.get(col, 'black') for col in df.columns]

# 设置绘图大小
plt.figure(figsize=(12, 6))

# 绘制热力图（vmin/vmax控制颜色对称）
ax = sns.heatmap(df, annot=True, fmt=".2f", cmap='RdBu_r', center=0,
                 linewidths=0.5, cbar_kws={"label": "相对基准GDP变化（%）"})

# 设置国家名称颜色
ax.set_xticklabels(df.columns, rotation=45, ha='right', fontsize=11)
for ticklabel, color in zip(ax.get_xticklabels(), col_colors):
    ticklabel.set_color(color)

# 设置标题和坐标
plt.title('')
plt.xlabel('')
plt.ylabel('')

plt.tight_layout()
plt.savefig('heatmap_gdp_2040.png', dpi=600, bbox_inches='tight')
plt.show()
```
