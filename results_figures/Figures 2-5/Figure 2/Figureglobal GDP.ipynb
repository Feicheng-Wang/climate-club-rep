```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# 设置中文字体为 Simsun（宋体）
plt.rcParams['font.sans-serif'] = ['Simsun']
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 读取 Excel 文件
file_path = 'clubresults0322.xlsx'
sheet_name = 'GLOBAL GDP'

# 从 Excel 中读取数据（从第1列第2行开始）
df = pd.read_excel(file_path, sheet_name=sheet_name, header=1, usecols="A:Y", nrows=7)

# 提取年份列表（列名，从第2列开始）
years = df.columns[1:]  # B1 到 Y1：2017 到 2040

# 绘图
plt.figure(figsize=(10, 6))
for i in range(len(df)):
    scenario = df.iloc[i, 0]          # 场景名称（A列）
    gdp_values = df.iloc[i, 1:]       # 对应每年的 GDP 数值
    plt.plot(years, gdp_values, marker='o', label=scenario)

# 图形美化
plt.title('不同Scenario下的全球GDP变化趋势')
plt.xlabel('年份')
plt.ylabel('全球GDP（单位可自定义）')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体和Seaborn全局风格
sns.set(style='whitegrid', font='Simsun', font_scale=1.2)
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 文件路径和sheet名称
file_path = 'clubresults0322.xlsx'
sheet_name = 'GLOBAL GDP'

# 读取年份行（B1:Y1）作为横轴
year_row = pd.read_excel(file_path, sheet_name=sheet_name, header=None, usecols="B:Y", nrows=1)
years = year_row.iloc[0].tolist()

# 读取 A2:Y 以获取所有 scenario 数据
df = pd.read_excel(file_path, sheet_name=sheet_name, header=1, usecols="A:Y")

# 设置颜色配色方案（Set2，柔和易区分）
palette = sns.color_palette("Set2", n_colors=len(df))

# 开始绘图
plt.figure(figsize=(10, 6))

for i in range(len(df)):
    scenario = df.iloc[i, 0]
    gdp_values = df.iloc[i, 1:].values
    if pd.isnull(scenario):
        continue
    plt.plot(years, gdp_values, marker='o', color=palette[i], linewidth=2)
    # 每条线末尾标注 scenario 名称（颜色与线一致）
    plt.text(years[-1] + 0.5, gdp_values[-1], scenario, fontsize=11, va='center', color=palette[i])

# 图表设置
plt.title('不同Scenario下的全球GDP变化趋势')
plt.xlabel('年份')
plt.ylabel('全球GDP相对基准情景变化（%）')

# 横坐标从2020开始，每隔3年显示一次
xticks = [year for year in years if year >= 2020][::3]
plt.xticks(xticks)

# 横坐标延伸到2045，为右侧留空间
plt.xlim(min(years), 2043)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体和风格
sns.set(style='whitegrid', font='Simsun', font_scale=1.2)
plt.rcParams['axes.unicode_minus'] = False

# 文件路径
file_path = 'clubresults0322.xlsx'

# 统一 scenario 顺序和颜色映射
scenario_order = ['CLUB0', 'CLUB1', 'CLUB2', 'CLUB3', 'CLUB0rt', 'CLUB1rt', 'CLUB2rt', 'CLUB3rt']
colors = sns.color_palette("Set2", n_colors=len(scenario_order))
scenario_colors = dict(zip(scenario_order, colors))

# --------------------------
# 辅助函数：绘制单个子图
# --------------------------
def plot_scenario(ax, sheet_name, y_label):
    # 读取年份
    year_row = pd.read_excel(file_path, sheet_name=sheet_name, header=None, usecols="B:Y", nrows=1)
    years = year_row.iloc[0].tolist()
    
    # 读取数据行
    df = pd.read_excel(file_path, sheet_name=sheet_name, usecols="A:Y", nrows=8)

    # 画图
    for i in range(len(df)):
        scenario = str(df.iloc[i, 0])
        values = df.iloc[i, 1:].values
        if pd.isnull(scenario):
            continue
        color = scenario_colors.get(scenario, 'gray')
        ax.plot(years, values, marker='o', linewidth=2, color=color)
        ax.text(years[-1] + 0.3, values[-1], scenario, fontsize=10, va='center', color=color)
    
    ax.set_xlabel('年份')
    ax.set_ylabel(y_label)
    ax.set_xticks(list(range(2020, 2045, 5)))
    ax.set_xlim(2015, 2043)
    ax.grid(True)

# --------------------------
# 创建并排图形
# --------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))  # 1行2列

plot_scenario(ax1, 'GLOBAL GDP', '全球GDP相对基准情景变化（%）')
plot_scenario(ax2, 'GLOBAL CO2', '全球CO2相对基准情景变化（%）')

plt.tight_layout()
plt.savefig('gdp_co2_scenario_combined.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 设置中文字体和图表风格
plt.rcParams['font.sans-serif'] = ['Simsun']
plt.rcParams['axes.unicode_minus'] = False

# 文件路径
file_path = 'clubresults0322.xlsx'

# 统一情景顺序和颜色（Tableau风格配色）
scenario_order = ['CLUB0', 'CLUB1', 'CLUB2', 'CLUB3', 'CLUB0rt', 'CLUB1rt', 'CLUB2rt', 'CLUB3rt']
colors = [
    "#1f77b4", "#ff7f0e", "#2ca02c", "#d62728",
    "#9467bd", "#8c564b", "#e377c2", "#7f7f7f"
]
scenario_colors = dict(zip(scenario_order, colors))

# ✨ 辅助函数：绘制单个子图
def plot_scenario(ax, sheet_name, y_label, offset_dict=None):
    # 读取年份
    year_row = pd.read_excel(file_path, sheet_name=sheet_name, header=None, usecols="B:Y", nrows=1)
    years = year_row.iloc[0].tolist()

    # 读取数据
    df = pd.read_excel(file_path, sheet_name=sheet_name, usecols="A:Y", nrows=8)

    for i in range(len(df)):
        scenario = str(df.iloc[i, 0])
        values = df.iloc[i, 1:].values
        if pd.isnull(scenario):
            continue
        color = scenario_colors.get(scenario, 'gray')
        ax.plot(years, values, marker='o', markersize=4, linewidth=2.5, color=color)
        
        # 添加文字标注（黑色） + 垂直偏移
        y_offset = 0
        if offset_dict and scenario in offset_dict:
            y_offset = offset_dict[scenario]
        
        ax.text(years[-1] + 0.5, values[-1] + y_offset, scenario,
                fontsize=12, va='center', color='black')

    ax.set_xlabel('年份', fontsize=12)
    ax.set_ylabel(y_label, fontsize=12)
    ax.set_xticks(list(range(2020, 2045, 5)))
    ax.set_xlim(2015, 2043)
    ax.grid(True)

# 🎯 CO2 图中的标注偏移设置（其他图不需要）
co2_label_offset = {
    'CLUB0': 0.4,
    'CLUB0rt': -0.3,
    'CLUB1': 0.1,
    'CLUB1rt': -0.4,
    'CLUB2': 0.2
}

# 🎨 创建并排图形
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

plot_scenario(ax1, 'GLOBAL GDP', '全球GDP相对基准情景变化（%）')                     # GDP图不调偏移
plot_scenario(ax2, 'GLOBAL CO2', '全球CO2相对基准情景变化（%）', co2_label_offset)  # CO2图指定错位

plt.tight_layout()
plt.savefig('gdp_co2_scenario.png', dpi=800, bbox_inches='tight')  # PNG高清图
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ✅ 设置全局字体为 Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号

# 文件路径
file_path = 'clubresults0322.xlsx'

# 情景顺序与配色
scenario_order = ['CLUB0', 'CLUB1', 'CLUB2', 'CLUB3', 'CLUB0rt', 'CLUB1rt', 'CLUB2rt', 'CLUB3rt']
colors = [
    "#1f77b4", "#ff7f0e", "#2ca02c", "#d62728",
    "#9467bd", "#8c564b", "#e377c2", "#7f7f7f"
]
scenario_colors = dict(zip(scenario_order, colors))

# ✨ 辅助函数：绘制单个子图
def plot_scenario(ax, sheet_name, y_label, offset_dict=None):
    # 年份行
    year_row = pd.read_excel(file_path, sheet_name=sheet_name, header=None, usecols="B:Y", nrows=1)
    years = year_row.iloc[0].tolist()

    # 数据行
    df = pd.read_excel(file_path, sheet_name=sheet_name, usecols="A:Y", nrows=8)

    for i in range(len(df)):
        scenario = str(df.iloc[i, 0])
        values = df.iloc[i, 1:].values
        if pd.isnull(scenario):
            continue
        color = scenario_colors.get(scenario, 'gray')
        ax.plot(years, values, marker='o', markersize=4, linewidth=2.5, color=color)

        # 添加情景名标签（黑色）+ 可选偏移
        y_offset = offset_dict.get(scenario, 0) if offset_dict else 0
        ax.text(years[-1] + 0.5, values[-1] + y_offset, scenario,
        fontsize=12, va='center', color='black', fontname='Times New Roman')


    # 设置横轴
    ax.set_xlabel('年份', fontsize=12)
    ax.set_xlim(2015, 2044)
    ax.set_xticks(list(range(2020, 2045, 5)))
    ax.grid(True)

    # 设置纵轴中文 → 指定 Simsun 字体
    ylabel = ax.set_ylabel(y_label, fontsize=14, fontweight='bold')
    ylabel.set_fontname('Simsun')

    # 同理，横轴中文也设置 Simsun
    xlabel = ax.xaxis.get_label()
    xlabel.set_fontname('Simsun')

# 🎯 CO2 图中的情景标签偏移设置
co2_label_offset = {
    'CLUB0': 0.4,
    'CLUB0rt': -0.3,
    'CLUB1': 0.1,
    'CLUB1rt': -0.4,
    'CLUB2': 0.2
}

# 🎨 并排图形
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

plot_scenario(ax1, 'GLOBAL GDP', '全球GDP相对基准情景变化（%）')
plot_scenario(ax2, 'GLOBAL CO2', '全球CO2相对基准情景变化（%）', co2_label_offset)

plt.tight_layout()
plt.savefig('gdp_co2_scenario.svg', bbox_inches='tight')
plt.show()
```
